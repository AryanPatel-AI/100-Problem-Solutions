```python
# 3. Input two integers and print their sum.
X = int(input("Enter the first number: "))
Y = int(input("Enter the second number: "))
print(f"The sum of {X} and {Y} is {X + Y}.")
```

```python
# 5. Convert temperature from Celsius to Fahrenheit.
celsius =float(input("Enter The Temperature In Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

```python
# 6. Swap two variables without using a third variable.
a = input("Enter the first variable: ")
b = input("Enter the second variable: ")
a, b = b, a
print(f"After swapping: a = {a}, b = {b}")
```

```python
# 7. Calculate the area of a circle given the radius.
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius ** 2
print(f"The area of the circle with radius {radius} is {area:.2f}.")
```

```python
# 8. Input a three-digit number and print the sum of its digits.
number = input("Enter a three-digit number: ")
if len(number) == 3 and number.isdigit():
    digit_sum = sum(int(digit) for digit in number)
    print(f"The sum of the digits in {number} is {digit_sum}.")
else:
    print("Please enter a valid three-digit number.")
```

```python
# 9. Convert minutes into hours and remaining minutes.
minutes = int(input("Enter the number of minutes: "))
hours = minutes // 60
remaining_minutes = minutes % 60
print(f"{minutes} minutes is equal to {hours} hours and {remaining_minutes} minutes.")
```

```python
# 10. Calculate simple interest (P, R, T given).
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest: "))
time = float(input("Enter the time period in years: "))
simple_interest = (principal * rate * time) / 100
print(f"The simple interest is {simple_interest}.")
```

```python
# 11. Convert kilometers to miles.
kilometers = float(input("Enter the distance in kilometers: "))
miles = kilometers * 0.621371
print(f"{kilometers} kilometers is equal to {miles} miles.")
```

```python
# 12. Print ASCII value of a character.
character = input("Enter a character: ")
if len(character) == 1:
    ascii_value = ord(character)
    print(f"The ASCII value of '{character}' is {ascii_value}.")
else:
    print("Please enter a single character.")
```

```python
# 13. Convert a number into binary format.
number = int(input("Enter a number: "))
binary = bin(number)[2:]  # Convert to binary and remove the '0b'
print(f"The binary representation of {number} is {binary}.")
```

```python
# 14. Input a float and round it to two decimal places.
float_num = float(input("Enter a float number: "))
rounded_num = round(float_num, 2)
print(f"The number {float_num} rounded to two decimal places is {rounded_num}.")
```

```python
# 15. Find the square and cube of a number.
number = float(input("Enter a number: "))
square = number ** 2
cube = number ** 3
print(f"The square of {number} is {square}.")
print(f"The cube of {number} is {cube}.")
```

# Module 2: Conditional Statements

```python
# 16. Check whether a number is even or odd.
number = int(input("Enter a number: "))
if number % 2 == 0:
    print(f"{number} is an even number.")
else:
    print(f"{number} is an odd number.")
```

```python
# 17. Check whether a number is positive, negative, or zero.
number = float(input("Enter a number: "))
if number > 0:
    print(f"{number} is a positive number.")
elif number < 0:
    print(f"{number} is a negative number.")
else:
    print(f"{number} is zero.")
```

```python
# 18. Find the largest among three numbers.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))
largest = max(num1, num2, num3)
print(f"The largest among {num1}, {num2}, and {num3} is {largest}.")
```

```python
# 19. Check if a year is a leap year.
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
# 20. Determine whether a character is a vowel or consonant.
char = input("Enter a character: ").lower()
if char in 'aeiou':
    print(f"{char} is a vowel.")
else:
    print(f"{char} is a consonant.")
```

```python
# 21. Check if a number is divisible by both 3 and 5.
number = int(input("Enter a number: "))
if number % 3 == 0 and number % 5 == 0:
    print(f"{number} is divisible by both 3 and 5.")
else:
    print(f"{number} is not divisible by both 3 and 5.")
```

```python
# 22. Grade calculator (marks to grade conversion).
marks = float(input("Enter the marks: "))
if marks >= 90:
    grade = 'A'
elif marks >= 80:
    grade = 'B'
elif marks >= 70:
    grade = 'C'
elif marks >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f"The grade for marks {marks} is {grade}.")
```

```python
# 23. Check if a number is a palindrome.
number = input("Enter a number: ")
if number == number[::-1]:
    print(f"{number} is a palindrome.")
```

```python
# 24. Check whether a triangle is valid (sum of angles).
angle1 = float(input("Enter the first angle of the triangle: "))
angle2 = float(input("Enter the second angle of the triangle: "))
angle3 = float(input("Enter the third angle of the triangle: "))
if angle1 + angle2 + angle3 == 180:
    print("The triangle is valid.")
else:
    print("The triangle is not valid.")
```

```python
# 25. Identify type of triangle (equilateral, isosceles, scalene).
side1 = float(input("Enter the length of the first side: "))
side2 = float(input("Enter the length of the second side: "))
side3 = float(input("Enter the length of the third side: "))
if side1 == side2 == side3:
    print("The triangle is equilateral.")
elif side1 == side2 or side2 == side3 or side1 == side3:
    print("The triangle is isosceles.")
else:
    print("The triangle is scalene.")
```

```python
# 26. Check if a number lies within a given range.
number = float(input("Enter a number: "))
lower_bound = float(input("Enter the lower bound of the range: "))
upper_bound = float(input("Enter the upper bound of the range: "))
if lower_bound <= number <= upper_bound:
    print(f"{number} lies within the range {lower_bound} to {upper_bound}.")
else:
    print(f"{number} does not lie within the range {lower_bound} to {upper_bound}.")
```
